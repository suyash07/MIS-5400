_Analysis by: Jack Kiefer_  
  

Since each player's birth state is provided by the dataset, we can correlate this to home-run averages in attempt and see which state manufactures the best babies that will grow up to be good at hitting home runs.

We first limit our analysis to players born in the U.S. The metric used for measuring ability to hit home runs is simply a home-run per at-bat percentage (HR/AB). ### Do batters with higher BMI get hit by the pitch more often?
_Analysis by: Jack Kiefer_

This analysis is especially pertinent to the goals of the report, as we find it amusing to think about batters missing the pitch and being struck by high-speed baseballs in unfortunate places.

The metric used for proclivity to being hit by the pitch is simply a hit-by-pitch per at-bat percentage (HBP/AB). We again sum these values over each player's entire careers before computing the percentage. We limit our analysis to players who have been to bat at least 100 times in order to avoid outliers, and we additionally only analyze players that have been hit by the pitch at least once.

The dataset provides height and weight in U.S. Customary, so we calculate each player's BMI according to the following formula:

$$\text{BMI} = \frac{\text{weight}_\text{lbs}}{\text{height}^2_{\text{in}}} \times 703$$

Finally, the correlation between each player's BMI and HBP percentage are visualized with a scatter plot and a bar chart.

## Which ten players have the highest batting averages across all teams?

### Steps :

1.  Read the csv Data.
2.  Perform various operations to carry out a meaningful dataframe
3.  Customize the dataframe to extract valuable visualization that estimates the result

```python
#Batting Average:no of Hits(H)/AB(At Bats)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import seaborn as sns
people = pd.read_csv('core/People.csv',usecols=['playerID','nameFirst','nameLast'])
#display(people.head(1))
bat= pd.read_csv('core/Batting.csv',usecols=['playerID','teamID','AB','H'])
#bat.head(1)
mergedPD = pd.merge(people,bat, how='inner', on='playerID')
mergedPD.drop(mergedPD[mergedPD.AB == 0].index,inplace=True)
mergedPD['Batting_AVG'] = (mergedPD['H']/mergedPD['AB'])*100
mergedPD['FullName']= mergedPD['nameFirst'] +' ' +  mergedPD['nameLast']
mergedPD.drop(columns = ['nameFirst','nameLast'],axis=0,inplace=True)
k=pd.DataFrame()
for key,value in mergedPD.groupby('teamID'):
    value = value.reset_index(drop=True).sort_values(by=['Batting_AVG'],ascending=False)
    valueOne = value.drop(value[value.AB < 40].index)[0:2]
    k= pd.concat([valueOne,k],axis=0)
k=k.sort_values(by=['Batting_AVG'],ascending=False).reset_index(drop=True)
k= k.reset_index(drop=True)[0:10]
display(k)
plt.figure(figsize=(20,15))
i=0
#colors=["red", "darkgrey", "#00FF7F",'#7CFC00','#FFFACD','#FF4500','#7CFC00','#601CD5','#00FFFF','#FF00FF','#C0C0C0']
labels=k['teamID']
while i <10:
    plt.bar(k['FullName'][i],k['Batting_AVG'][i],label = labels[i],width=0.5)
    i+=1
plt.xticks(k['FullName'][0:10],fontsize=15,rotation=45)
sns.set_palette("RdYlBu")
plt.legend(ncol=3,fontsize=15)
plt.xlabel('Names_of_Players',fontsize=15)
plt.ylabel('Batting_Average(H/AB)',fontsize=15)
plt.title('Batters\'_Name VS Top_Ten_Batting_Averages_of_all_teams',fontsize=15)
plt.tick_params(labelleft=False,left=False)
plt.box(False)
for i ,value in enumerate(k['Batting_AVG'][0:10]):
    plt.annotate(str(round(value,2)),xy=(i,value+0.5),ha='center',fontsize=15)
plt.savefig('BarPlot.png')
plt.show()
```

# What are the highest and lowest batting averages of top ten players of each team?

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import seaborn as sns
people = pd.read_csv('core/People.csv',usecols=['playerID','nameFirst','nameLast'])
#display(people.head(1))
bat= pd.read_csv('core/Batting.csv',usecols=['playerID','teamID','AB','H'])
#bat.head(1)
mergedPD = pd.merge(people,bat, how='inner', on='playerID')
mergedPD.drop(mergedPD[mergedPD.AB == 0].index,inplace=True)
mergedPD['Batting_AVG'] = (mergedPD['H']/mergedPD['AB'])*100
mergedPD['FullName']= mergedPD['nameFirst'] +' ' +  mergedPD['nameLast']
mergedPD.drop(columns = ['nameFirst','nameLast'],axis=0,inplace=True)
v=pd.DataFrame()
for key,value in mergedPD.groupby('teamID'):
    value = value.reset_index(drop=True).sort_values(by=['Batting_AVG'],ascending=False)
    valueOne = value.drop(value[value.AB < 40].index)[0:10]
    valueOne=valueOne.sort_values(by=['Batting_AVG'],ascending=False).reset_index(drop=True)
    valueOne= valueOne.reset_index(drop=True)
    v = pd.concat([v,valueOne],axis=0)
#display(v)
v= v[0:99]
#display(v)
plt.figure(figsize=(20,15))
sns.lineplot(x=v['FullName'],y=v['Batting_AVG'],hue=v['teamID'],palette='cubehelix',linewidth=10.0)
legend = plt.legend(ncol=3,fontsize=30)
plt.xlabel('')
plt.ylabel('Batting_Average(H/AB)',fontsize=15)
plt.title('Highest_and_lowest_Batting_Averages_of_Top_10_Players_of_each_team',fontsize=25)
plt.tick_params(labelbottom=False,bottom=False)
legend.get_frame().set_linewidth(10)
for line in legend.get_lines():
    line.set_linewidth(10.0)
#plt.box(False)
plt.savefig('Pic_2.png')
plt.show()
```

## Which state produces the best homerun-hitters?

First, read in the data

```python
import pandas as pd
import plotly

players = pd.read_csv('People.csv')
players = players[players.birthCountry == 'USA' ]
players = players[['playerID', 'nameFirst', 'nameLast','birthState']]

# Read in batting stats
battingStats = pd.read_csv('Batting.csv')
battingStats = battingStats[['playerID','AB','HR']]
```

Next, operate on the data.

```python
# Get total sum of AB and HR over players' entire careers
battingStats = battingStats.groupby('playerID').sum()

# Join the batting stats and player data
stats = players.merge(battingStats, on='playerID')

# Drop any gross null values!
stats = stats.dropna() 

# No LUCKY SMALL-TIME CHUMS.
# Only analyze players that have been to bat at least 100 times.
stats = stats[stats.AB >= 100]

# Calculate atbats per homerun given a row of player data
def homerunsPerAtBat(row):
    return row['HR'] / row['AB']

# Compute the HR/AB for each entry and add it to the dataframe
stats['HRpAB'] = stats.apply(lambda row: homerunsPerAtBat(row), axis=1)

# Group players by their birth state and take the mean!
stats = stats.groupby('birthState').mean().reset_index()

# Uncomment these lines to display a ranked list of all states
#stats = stats.sort_values(['HRpAB'],ascending=False)
display(stats)
```

Finally, produce a pretty visualization.

```python
# Format data for choropleth visualization
data = [ dict(
        type='choropleth',
        autocolorscale=True,
        locations = stats['birthState'],
        z = stats['HRpAB'].astype(float),
        locationmode = 'USA-states',
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Mean HR / AB")
        ) ]

# Define some viz properties
layout = dict(
        title = 'All-time Mean Homeruns per At-Bat of American Players by Birth State',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(0, 10, 255)'),
             )
    
# Kachow!
fig = dict( data=data, layout=layout )
#plotly.offline.init_notebook_mode()
plotly.offline.iplot( fig)
'''
'equirectangular', 'mercator', 'orthographic', 'natural
            earth', 'kavrayskiy7', 'miller', 'robinson', 'eckert4',
            'azimuthal equal area', 'azimuthal equidistant', 'conic
            equal area', 'conic conformal', 'conic equidistant',
            'gnomonic', 'stereographic', 'mollweide', 'hammer',
            'transverse mercator', 'albers usa', 'winkel tripel',
            'aitoff', 'sinusoidal']
'''
```

## Do batters with higher BMI get hit by the pitch more often?

Read in data.

```python
players = pd.read_csv('People.csv')
players = players[['playerID', 'height', 'weight']]

# Read in hit-by-the-pitch and at-bat stats
HBPstats = pd.read_csv('Batting.csv')
HBPstats = HBPstats[['playerID','HBP', 'AB']]
```

Operate on data.

```python
# Get total sum of HBP over players' entire careers
HBPstats = HBPstats.groupby('playerID').sum()

# Join the HBP stats and player data
stats = players.merge(HBPstats, on='playerID')

# Drop any gross null values!
stats = stats.dropna() 

# Only look at people who have been hit by the pitch at least once and have
# been to bat at least 100 times
stats = stats[stats.AB >= 100]
stats = stats[stats.HBP >= 1]
```

The database provides height and weight in inches and pounds, respectively. Thus, each player's BMI is given by:
$$\text{BMI} = \frac{\text{weight}_\text{lbs}}{\text{height}^2_{\text{in}}} \times 703$$


```python
def getBMI(row):
    return (row['weight'] / (row['height']**2))*703

def HBPpercent(row):
    return row['HBP'] / row['AB']

# Compute the BMI for each entry and add it to the dataframe
stats['BMI'] = stats.apply(lambda row: getBMI(row), axis=1)
stats['HBPpercent'] = stats.apply(lambda row: HBPpercent(row), axis=1)

# Have a look at bucketized BMIs and average HBP for those BMIs
BMIavgs = stats.copy()
BMIavgs.BMI = BMIavgs.BMI.round()
BMIavgs = BMIavgs.groupby('BMI').mean().reset_index()
```

Produce pretty visualizations!

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(stats['BMI'],stats['HBPpercent'])
plt.title('BMI vs. Hit-by-pitch percentage')
plt.xlabel('BMI')
plt.ylabel('Hit-by-pitch per at-bat')
plt.grid()
plt.savefig('BMIScatters.png')
plt.show()

plt.figure()
plt.title('BMI vs. Average HBP Percent')
plt.xlabel('BMI (discretized)')
plt.ylabel('Average hit-by-pitch percent')
plt.grid(False)
plt.bar(BMIavgs['BMI'],BMIavgs['HBPpercent'])
plt.savefig('BMIbars.png')
plt.show()
```
